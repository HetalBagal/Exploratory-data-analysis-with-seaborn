```python
import seaborn as sns
```

```python
#Basic ploting using seaborn

tips=sns.load_dataset('tips')
tips
```

```python
#Create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
# Line Plot

sns.lineplot(x='size', y='total_bill', data= tips)
plt.title('Line Plot of Total bill by size')
plt.show()
```

```python
#Categporical plots
# Bar Plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot of Total bills by Days')
plt.show()
```

```python
# Box Plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Box Plot of Total bill in Day")
plt.show()
```

```python
# Violin Plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
#Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
#Kde plot(Kernel Density Estimater)
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
#Pairplot

sns.pairplot(tips)
```

```python
# HeatMap
corr = tips[['total_bill' ,'tip', 'size']].corr()
corr

```

```python
sns.heatmap(corr, annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df= pd.read_csv('sales_data.csv')
sales_df.head(5)
```

```python
# Plot total sales by Customer Id
plt.figure(figsize=(10,6))
sns.barplot(x='Customer ID' , y='Total Amount', data=sales_df , estimator=sum)
plt.title('Total Sales by Customer ID')
plt.xlabel('Customer ID')
plt.ylabel('Total Amount')
plt.show()
```

```python
# Plot total sales by Product Category
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category' , y='Total Amount', data=sales_df , estimator=sum)
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount')
plt.show()
```
